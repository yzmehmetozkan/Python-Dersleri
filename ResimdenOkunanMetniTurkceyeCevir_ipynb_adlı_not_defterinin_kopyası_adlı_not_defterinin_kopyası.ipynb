```python
## ardından, metin çevirisi için googletrans kütüphanesini kullanarak önce Tesseract OCR ile elde edilen metni alıp çevirebiliriz. Örnek bir kod şu şekilde olabilir:
```

```python
from PIL import Image
import pytesseract
from googletrans import Translator
from google.colab import files

def resimden_metin_oku(resim_yolu):
    try:
        img = Image.open(resim_yolu)
        metin = pytesseract.image_to_string(img)
        return metin
    except Exception as e:
        return f"Hata: {e}"

def cevir(metin, hedef_dil='tr'):
    try:
        translator = Translator()
        ceviri = translator.translate(metin, dest=hedef_dil)
        return ceviri.text
    except Exception as e:
        return f"Hata: {e}"

if __name__ == "__main__":
    uploaded = files.upload()

    for resim in uploaded.keys():
        okunan_metin = resimden_metin_oku(resim)
        print(f"{resim} dosyasından okunan metin:")
        print(okunan_metin)

        cevrilmis_metin = cevir(okunan_metin)
        print("\nÇevrilen Metin:")
        print(cevrilmis_metin)
```

```python
from PIL import Image, ImageDraw, ImageFont

def metni_resme_ekle(metin, resim_yolu="yeni_resim.jpg"):
    # Yeni bir boş resim oluştur
    genislik, yukseklik = 600, 400
    arka_plan_rengi = (255, 255, 255)  # Beyaz arka plan rengi
    resim = Image.new("RGB", (genislik, yukseklik), arka_plan_rengi)

    # Metni resme yerleştir
    cizim = ImageDraw.Draw(resim)
    font = ImageFont.load_default()  # Varsayılan font kullanılabilir, isterseniz başka bir font da seçebilirsiniz
    metin_boyutu = cizim.textsize(metin, font=font)
    metin_konumu = ((genislik - metin_boyutu[0]) / 2, (yukseklik - metin_boyutu[1]) / 2)
    cizim.text(metin_konumu, metin, fill=(0, 0, 0), font=font)  # Metni siyah renkte yaz

    # Resmi kaydet
    resim.save(resim_yolu)

if __name__ == "__main__":
    cevrilmis_metin = "Bu metin Türkçe'dir."
    metni_resme_ekle(cevrilmis_metin, "yeni_resim.jpg")
```

```python
from PIL import Image, ImageDraw, ImageFont

def metni_resme_ekle(metin, resim_yolu="yeni_resim.jpg"):
    # Yeni bir boş resim oluştur
    genislik, yukseklik = 600, 400
    arka_plan_rengi = (255, 255, 255)  # Beyaz arka plan rengi
    resim = Image.new("RGB", (genislik, yukseklik), arka_plan_rengi)

    # Metni resme yerleştir
    cizim = ImageDraw.Draw(resim)
    font = ImageFont.load_default()  # Varsayılan font kullanılabilir, isterseniz başka bir font da seçebilirsiniz
    metin_boyutu = cizim.textbbox((0, 0), metin, font=font)
    metin_konumu = ((genislik - metin_boyutu[2]) / 2, (yukseklik - metin_boyutu[3]) / 2)
    cizim.text(metin_konumu, metin, fill=(0, 0, 0), font=font)  # Metni siyah renkte yaz

    # Resmi kaydet
    resim.save(resim_yolu)

if __name__ == "__main__":
    cevrilmis_metin = "Bu metin Türkçe'dir."
    metni_resme_ekle(cevrilmis_metin, "yeni_resim.jpg")
```
